a) Write the code to construct from scratch an **unidirected** random graph with 10 nodes using the Gilbert model, i.e., the probability of an edge existing between a pair of nodes is p. For this exercise use p = 0.4.

- *Note that you cannot use any of the functions that automatically creates a random graph from NetworkX library.*

````python
# Milestone 1
# a)

# Import the random library
import random

# Create an empty undirected graph
G_undirected_random = nx.Graph()

# Number of nodes
n = 10

# Probability of an edge
p = 0.4

# Add nodes to the graph
nodes = range(n)
G_undirected_random.add_nodes_from(nodes)

# Generate edges based on the probability 'p'
for i in range(n):
    for j in range(i + 1, n):
        if random.random() < p:
            G_undirected_random.add_edge(i, j)

# Plot the graph using spring layout
pos = nx.spring_layout(G_undirected_random)
nx.draw_networkx(G_undirected_random, pos)

# Add title and display the graph
plt.title("Undirected Random Graph")
plt.show()
````

````python
#Solution Milestone 2, question d)
# From milestone 1 b)
G_directed_random = nx.DiGraph() # Create an empty directed graph
n = 10 # Number of nodes
p = 0.4 # Probability of an edge
nodes = range(n)
G_directed_random.add_nodes_from(nodes) # Add nodes to the graph
# Generate directed edges based on the probability 'p'
for i in range(n):
    for j in range(n):
        if i != j and random.random() < p:
            G_directed_random.add_edge(i, j)

# Compute the Betweenness centrality manually without defining a function
betweenness_centrality_computed = {node: 0 for node in nodes}

for s in G_directed_random.nodes():
    stack = []
    predecessors = {node: [] for node in nodes}
    sigma = {node: 0 for node in nodes}
    sigma[s] = 1
    distance = {node: -1 for node in nodes}
    distance[s] = 0
    queue = [s]
    while queue:
        v = queue.pop(0)
        stack.append(v)
        for neighbor in G_directed_random.neighbors(v):
            if distance[neighbor] < 0:
                queue.append(neighbor)
                distance[neighbor] = distance[v] + 1
            if distance[neighbor] == distance[v] + 1:
                sigma[neighbor] += sigma[v]
                predecessors[neighbor].append(v)
    dependency = {node: 0 for node in nodes}
    while stack:
        w = stack.pop()
        for predecessor in predecessors[w]:
            dependency[predecessor] += (sigma[predecessor] / sigma[w]) * (1 + dependency[w])
            if w != s:
                betweenness_centrality_computed[w] += dependency[w]
betweenness_centrality_computed = {k: v / 2.0 for k, v in betweenness_centrality_computed.items()}

# Compute Betweenness centrality using NetworkX function
betweenness_centrality_nx = nx.betweenness_centrality(G_directed_random)

# Print and compare the results:
print("\n Comparison Betweenness Centrality (NetworkX and Computed):")
for node, betweenness_centrality in betweenness_centrality_nx.items():
    print("Node " + str(node) + ": Using NetworkX functions -> " + str(betweenness_centrality) + " AND computed -> " + str(betweenness_centrality_computed[node]))

````

d) Write a code that computes the Pagerank for all nodes in the graph. *Note the code cannot use the function defined for this purpose in the NetworkX library*
> Use a stop condition such that the sum of the Pagerank value of of all nodes in the graph for iteration i is < 10^-5 compared to iteration i-1.

Print the results of your solution along with the results obtained with the functions of NetworkX library to compute the Pagerank. **Note tha both, your solution and NetworkX functions should provide the same results.**



````python
#Solution Milestone 2
# e)
# From milestone 1 b)
G_directed_random = nx.DiGraph() # Create an empty directed graph
n = 10 # Number of nodes
p = 0.4 # Probability of an edge
nodes = range(n)
G_directed_random.add_nodes_from(nodes) # Add nodes to the graph
# Generate directed edges based on the probability 'p'
for i in range(n):
    for j in range(n):
        if i != j and random.random() < p:
            G_directed_random.add_edge(i, j)

# Compute Pagerank manually with a stop condition
damping_factor = 0.85 # Typically is set to 0.85
stop = 1e-5 # Stop condition

pagerank_computed = {node: 1 / n for node in G_directed_random.nodes()}
converged = False

while not converged:
    pagerank_next = {node: (1 - damping_factor) / n for node in G_directed_random.nodes()}
    for node in G_directed_random.nodes():
        for predecessor in G_directed_random.predecessors(node):
            out_degree = len(list(G_directed_random.successors(predecessor)))
            pagerank_next[node] += damping_factor * pagerank_computed[predecessor] / out_degree
    sum_diff = sum(abs(pagerank_next[node] - pagerank_computed[node]) for node in G_directed_random.nodes())
    if sum_diff < stop:
        converged = True
    pagerank_computed = pagerank_next

# Compute Pagerank using NetworkX function
pagerank_nx = nx.pagerank(G_directed_random, alpha=damping_factor)

# Print and compare the results:
print("\n Comparison PageRank Centrality (NetworkX and Computed):")
for node, pagerank_centrality in pagerank_nx.items():
    print("Node " + str(node) + ": Using NetworkX functions -> " + str(pagerank_centrality) + " AND computed -> " + str(pagerank_computed[node]))
````
