# **2.1. MILESTONE 1: POPULATION**:
Retrieve the 2022 countries' population and show the Top 10 countries and the Bottom 10 countries within the World Bank database.



```python
# Import the necessary libraries
import requests
import pandas as pd
```

After trying different links we found out which part is the codification indicator of the countries' population.

Also we used "api.worldbank.org" in the base URL for the World Bank's API. This way we will be able to make requests for a structured format JSON (and avoid the error: JSONDecodeError: Expecting value: line 1 column 1 (char 0))

```python
indicator = 'SP.POP.TOTL'
url = 'https://api.worldbank.org/v2/country/indicator/' + indicator + '?format=json&date=2022&per_page=20000'
```

Retrieve the data and convert it into a pandas dataframe so we have an easy access to it:

```python
response = requests.get(url)
data = response.json() # dictionary or a list of dictionaries
df = pd.DataFrame(data[1])
df
```

We added a new column with just the name of the countries so the printed results are clear and without any irrelevant data:

```python
# Get the country name out of the dictionary and create a new column
df['country_name'] = [country['value'] for country in df['country']]
df
```

At this point, we noticed that we were not getting the desired results by country but by aggregations or zones, so we decided to get the tag for each country.

```python
#query without indicator to get all countries
url = 'https://api.worldbank.org/v2/country/?format=json&date=2022&per_page=20000'
```

```python
response = requests.get(url)
data = response.json() # dictionary or a list of dictionaries
df_countries = pd.DataFrame(data[1])
df_countries
```

We noticed that in the *region* column the rows that are not countries but groups of countries, the value is *Aggregates*. So we removed those rows from our dataset to keep only countries.

```python
df_countries['region_name'] = [id['value'] for id in df_countries['region']] #this creates a column with the region value
#df_countries
f_df_countries = df_countries[df_countries['region_name'] != 'Aggregates'] #this filters our dataset only for countries and not aggregations
f_df_countries
```

Now we are going to store all the countries' labels or codes (which are in the *iso2Code* column) in a list, so that we can filter the first dataset we had (df) with the population information we want.

```python
codes = [] #list to store the countries' codes
for code in f_df_countries['iso2Code']:
  codes.append(code)

# codes
```

```python
df['country_code'] = [country['id'] for country in df['country']] #first we add a column with the codes
#df
final_df = df[df['country_code'].isin(codes)] #filter by the codes corresponding to countries
final_df
```

Finally, we sorted and retrieved the top (bottom) ten results to show them in order:

```python
# Top 10 countries
top_10 = final_df.sort_values(by='value', ascending=False).head(10)

# Print the results (Top 10)
print("Top 10 Countries by Population in 2022:")
print(top_10[['country_name', 'value']])
```

```python
# Bottom 10 countries
bottom_10 = final_df.sort_values(by='value').head(10)

# Print the results (Bottom 10)
print("Bottom 10 Countries by Population in 2022:")
print(bottom_10[['country_name', 'value']])
```

# **2.2. MILESTONE 2: WOMEN Vs. MEN POPULATION**:
Obtain the % of men and women for each country and compute the difference among them using the formula %women - %men. Display:

1- The number of countries with more women than men.

2- The number of countries with more men than women

3- The 10 countries with more women compared to men (ten countries with the largest positive value of the previous metric)

- The 10 countries with more men compared to women (ten countries with the largest negative value of the previous metric).

**Note**: You can use the indicator the absolute number of men and women from the World Bank API and compute the % for each country and the difference, or you can use the indicator given directly the %.



#### Data request and structuring into data frame

To begin with, we got 2 separate datasets with the data corresponding to the Percentage of population of women and of men for each country. Later we found out both codification indicators for the percentage of total population of women and men, and we stored the results of the get in a json.

```python
indicator_women = 'SP.POP.TOTL.FE.ZS' #Population, female (% of total population)
indicator_men = 'SP.POP.TOTL.MA.ZS' #Population, male (% of total population)
url_women = 'https://api.worldbank.org/v2/country/indicator/' + indicator_women + '?format=json&per_page=20000&date=2022'
url_men = 'https://api.worldbank.org/v2/country/indicator/' + indicator_men + '?format=json&per_page=20000&date=2022'

# Send the GET requests to the World Bank API
response_women = requests.get(url_women)
response_men = requests.get(url_men)

#both are dictionary or a list of dictionaries
wdata = response_women.json()
mdata = response_men.json()

#convert the json to dataset
wdf = pd.DataFrame(wdata[1])
mdf = pd.DataFrame(mdata[1])
wdf
```

#### Filtering our data

Again we have to filter by the code to get only the countries and not groups, so we will use again the auxiliary list we created with the countries' codes.

```python
wdf['country_code'] = [country['id'] for country in wdf['country']] #first we add a column with the codes
mdf['country_code'] = [country['id'] for country in wdf['country']] #first we add a column with the codes
#wdf
wdf = wdf[wdf['country_code'].isin(codes)] #filter by the codes corresponding to countries
mdf = mdf[mdf['country_code'].isin(codes)] #filter by the codes corresponding to countries
wdf
```

#### Structuring and transforming our data

Then, we added a column with the name of each country to both datasets. Afterwards, we merged both datasets by mapping the country name and keeping the value for the % of men and the % of women. Then, we added a column with the difference between the percentage of women and the percentage of men.

```python
# Get the country name out of the dictionary and create a new column
wdf['country_name'] = [country['value'] for country in wdf['country']]
mdf['country_name'] = [country['value'] for country in mdf['country']]

wdf = wdf.rename(columns={'value': 'value_women'})
mdf = mdf.rename(columns={'value': 'value_men'})

#Merge both dataframes so that we can compute the operations with the values of both
df = pd.merge(wdf[['country_name', 'value_women','date']], mdf[['country_name', 'value_men','date']], on=['country_name','date'], how='inner')

#Create a new column with the diference: %women - %men
df['difference'] = df['value_women'] - df['value_men']

df  #preview of our data
```

#### Calculalting the desired results

Now that we have the data needed in our dataset we can calculate the number of countries with more women than men, more men than women and the top 10 countries of each one. This was easy thanks to the column we created with the difference in the percentages (i.e., the positive values of "difference" belong to the countries with more women than men, and the positive values to the countries with more men than women)


```python
#1- The number of countries with more women than men. (Difference column is possitive)
more_women = df[df['difference']>0]
print("The number of countries with more women than men is:", len(more_women))
#print(more_women)

#2- The number of countries with more men than women
more_men = df[df['difference']<0]
print("The number of countries with more men than women is:", len(more_men))

#3A- The 10 countries with more women compared to men (ten countries with the largest positive value of the previous metric)
top_10_women = more_women.sort_values(by='difference', ascending=False).head(10)

print("Top 10 countries with more women compared to men:")
print(top_10_women[['country_name', 'value_women','difference']])

#3B- The 10 countries with more men compared to women (ten countries with the largest negative value of the previous metric).
top_10_men = more_men.sort_values(by='difference', ascending=True).head(10)

print("Top 10 countries with more men compared to women:")
print(top_10_men[['country_name', 'value_men','difference']])
```

## **2.3. MILESTONE 3: GDP PER CAPITA ACCORDING FOR INCOME LEVEL GROUPS**:

Compute the average increase/decrease in percentage for the PIB per capita in US dollars in the following two periods: 2000-2022 and  2010-2022, for the following income groups: low-income economies, lower-middle-income economies, middle economies, upper-middle-income economies and high-income economies. The following, link provides information of the different country aggregations carried out by the World Bank.  

https://datahelpdesk.worldbank.org/knowledgebase/articles/906519-world-bank-country-and-lending-groups

 You should compute the %PIB increase as follows. Given country A with a PIB Per Capita \$20000 in 2000 and \$30000 in 2022 the increase/decrease should be computed as follow:

%PIB increase = 100*(30000-20000)/20000=50%.


#### Data request and structuring into data frame

We began by getting the dataset with the data corresponding to the PIB per capita (GDP) in US dollars, by finding out the codification indicator for it and storing the results of the get in a json.

```python
indicator = 'NY.GDP.PCAP.CD' #indicator code for GDP per capita (current US$)
url = 'https://api.worldbank.org/v2/country/all/indicator/' + indicator + '?format=json&per_page=20000'

response = requests.get(url) # Send the GET request to the World Bank API
data = response.json() # dictionary or a list of dictionaries (json format)
df = pd.DataFrame(data[1]) #convert to dataset
df #preview of the raw dataset
```

#### Year range filtering step and calculation of % of PIB increase

Now we will perform some data cleaning and filtering in the raw dataset to get a structured df with the information we want. We add a column with the name of each country and filter the date columns to keep only the years we are interested in (2000, 2010 and 2022). With this information we can calculate the % of PIB increase as indicated above. Lastly, we will delete values with na values, as they will not give any result for the % of increase.

```python
# Get the country name out of the dictionary and create a new column
df['country_name'] = [country['value'] for country in df['country']] #add column for the country name


df = df.pivot(index = 'country_name', columns = 'date', values = 'value')[['2000', '2010', '2022']] #keep only data for the years of interest

# Create columns with the percentage of increase in the PIB for the 2 desired ranges of years
df['% of PIB increase 2000 to 2022'] = 100*(df['2022']- df['2000'])/df['2000']
df['% of PIB increase 2010 to 2022'] = 100*(df['2022']- df['2010'])/df['2010']

df_years = df.dropna() #data cleaning

df_years
```

#### Income groups filtering step

In the previous step we noticed that the country column in our first dataframe already contained the information we wanted about the PIB income by income groups, so it was simpler for us. So we only had to keep those groups in order to have the desired results.

```python
groups = ['Low income','Lower middle income','Middle income','Upper middle income','High income'] #list with names of each income group as they appear in the dataframe above

groups_df = df_years[df_years.index.isin(groups)] #keep only the desired columns (the ones in the groups list)
groups_df = groups_df.rename_axis('Income groups') #change name of index
groups_df
```

# **2.4. MILESTONE 4: TOP 5 COUNTRIES INCREASE GDP PER INCOME-GROUP**

For each of the income groups included in Milestone 3 and the period 2010-2022 list the Top 5 countries in terms of %GDPR per capita increase along with the value

**NOTE**: Do not consider the countries for which you do not have data either in 2010 or 2022 or both of them

```python
import requests
import pandas as pd

# Income levels: 'High income'--> 'HIC'/ 'Low income'--> 'LIC'/ 'Lower middle income'--> 'LMC'/ 'Middle income'--> No entries/ 'Upper middle income'-->'UMC'
incomeLevels = ['HIC', 'LIC', 'LMC', 'UMC']
countries = {}

for i in incomeLevels:
  url = f'http://api.worldbank.org/V2/country?incomeLevel={i}&format=json&per_page=1000'
  response = requests.get(url)
  data = response.json()[1]

  df = pd.DataFrame(data)
  countries[i] = list(df['id'])

  countryNames = []
  gdp2010 = []
  gdp2022 = []

  for id in countries[i]:
    url2010 = f'https://api.worldbank.org/v2/country/{id}/indicators/' + indicator + '?format=json&date=2010&per_page=4000'
    url2022 = f'https://api.worldbank.org/v2/country/{id}/indicators/' + indicator + '?format=json&date=2022&per_page=4000'

    response2010 = requests.get(url2010)
    data2010 = response2010.json()[1]

    response2022 = requests.get(url2022)
    data2022 = response2022.json()[1]
    if data2010 is not None:
      countryNames.append(data2010[0]['country']['value'])
      gdp2010.append(data2010[0]['value'])
      gdp2022.append(data2022[0]['value'])

  income = pd.DataFrame({'Country': countryNames, 'GDP2010': gdp2010, 'GDP2022': gdp2022})

  gdpIncrease = []

  for idx, row in income.iterrows():
    gdpIncrease.append(100*(row[2]-row[1]/row[1]))

  income['Increase-GDP'] = gdpIncrease

  top5_group = income.nlargest(5, 'Increase-GDP')
  print(i,'level:')
  print(top5_group)
```

# **2.5. MILESTONE 5: CO2 emission per capita**

Retrieve the most recent non empty value for the amount of CO2 emission per capita (metric tons per country) for all the countries. Display the 30 countries with the highest CO2 emission per capita along with their value and the year related to that value.

**NOTE**: You cannot search manually the year and use it in your query for this milestone.


#### Data request and structuring into data frame
We did the same process than in the previous milestones to get the data we wanted, including the step of adding a column with the country name and delete the empty data.

```python
indicator = 'EN.ATM.CO2E.PC' #CO2 emissions (metric tons per capita)
url = 'https://api.worldbank.org/v2/country/indicator/' + indicator + '?format=json&per_page=20000'
response = requests.get(url)
data = response.json() # dictionary or a list of dictionaries
df = pd.DataFrame(data[1])
df['country_name'] = [country['value'] for country in df['country']] #add column with the country name
df = df.dropna() #delete empty data

df #preview of our data
```

```python
df['country_code'] = [country['id'] for country in df['country']] #first we add a column with the codes
df = df[df['country_code'].isin(codes)] #filter by the codes corresponding to countries
df
```

#### Select the most recent value for each country in a non-manual way

To keep the most recent row (latest date) for each country in our dataset, we first sorted our data by the 'date' column in descending order and then we eliminated the duplicates by 'country_name', so that we could keep only the most recent value for each country.


```python
# Sort the dataset by the 'date' to get the most recent year
df.sort_values(by='date', ascending=False, inplace=True)

#Eliminate duplicate values keeping only the first row of each country ( so it is the most recent value)
df = df.drop_duplicates(subset='country_name', keep='first')

df #preview of our data
```

Then we got the top 30 countries by the value of CO2 emission by sorting our data by the column value:

```python
#Top 30
top_30 = df.sort_values(by='value', ascending=False).head(30)

# Print the results (Top 30)
print("Top 30 countries with the highest CO2 emission per capita:")
print(top_30[['country_name','value','date']])
```
